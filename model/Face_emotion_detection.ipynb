```python
#General
import os
import matplotlib.pyplot as plt
import numpy as np
import random

#TF
import tensorflow as tf
from tensorflow import keras
from keras import regularizers
from tensorflow.keras import layers
#CV
import cv2
```

```python
# Vars
Classes=['angry','disgust','fear','happy','neutral','sad','surprise']
emotion_labels = {
                0: "Angry",
                1: "Disgust",
                2: "Fear",
                3: "Happy",
                4: "neutral",
                5: "sad",
                6: "surprise"
                }
img_size=224 # for tranfer learing
np.random.seed(42)
max_images_per_folder = 3000
batch_size = 128
```

```python
# Printing an imag from each cat
for category in Classes:
    path=os.path.join("train",category)
    for img in os.listdir(path):
        print(category)
        img_array=cv2.imread(os.path.join(path,img))
        #plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        plt.imshow(img_array)
        plt.show()
        break
```

# Converting all the imgs to array

```python
def create_training_data():
    training_data = []
    for category in Classes:
        path = os.path.join("train", category)
        class_num = Classes.index(category)  # 0=angry, 1=happy, 2=sad, 3=neutral
        image_counter = 0
        for img in os.listdir(path):
            if image_counter >= max_images_per_folder:
                break
            try:
                img_path = os.path.join(path, img)
                img_array = cv2.imread(img_path)
                img_array = cv2.resize(img_array, (img_size, img_size))
                img_array = img_array.astype(np.float32) / 255.0
                training_data.append((img_array, class_num))
                image_counter += 1
            except Exception as e:
                pass
    return training_data

training_data = create_training_data()

# Shuffle training data with seed 42
np.random.shuffle(training_data)

# Split into X and y
X = [item[0] for item in training_data]
y = [item[1] for item in training_data]

# Convert X to a numpy array
X = np.array(X)
y = np.array(y)

# Create TensorFlow Dataset from X and y
dataset = tf.data.Dataset.from_tensor_slices((X, y))

# Shuffle, batch, prefetch, and cache the dataset
dataset = dataset.shuffle(len(training_data), seed=42)
dataset = dataset.batch(batch_size)
dataset = dataset.prefetch(tf.data.AUTOTUNE)
dataset = dataset.cache()

print(len(dataset))
```

# Model

```python
base_model = tf.keras.applications.MobileNetV3Small()
```

```python
base_input=base_model.layers[0].input
base_output=base_model.layers[-2].output
final_output=layers.Dense(128)(base_output) 
final_output=layers.Activation('relu')(final_output)
final_output = layers.Dropout(0.5)(final_output)  
final_output=layers.Dense(64)(final_output)
final_output=layers.Activation('relu')(final_output)
final_output = layers.Dropout(0.5)(final_output)  

final_output=layers.Dense(7,activation='softmax', kernel_regularizer=regularizers.l2(0.01))(final_output) # I have 7 classes
my_model=keras.Model(inputs=base_input,outputs=final_output)
callback = keras.callbacks.EarlyStopping(monitor='loss', patience=3)
my_model.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
my_model.summary()
```

```python
my_model.fit(dataset,epochs=50)
```

```python
#my_model.save('FeelingsDetector.h5')
my_model = keras.models.load_model('FeelingsDetector.h5')
face_cascade = cv2.CascadeClassifier('opencv/haarcascades/haarcascade_frontalface_default.xml')
eye_cascade = cv2.CascadeClassifier('opencv/haarcascades/haarcascade_eye.xml')
```

```python
import numpy as np

training_data = []
total_images = 0
correct_predictions = 0

def create_testing_data():
    max_images_per_folder = 300  # Setting limit for memory issues
    global total_images, correct_predictions
    
    for category in Classes:
        path = os.path.join("train", category)
        class_num = Classes.index(category)  # 0=angry, etc.
        image_counter = 0
        
        for img in os.listdir(path):
            if image_counter >= max_images_per_folder:
                break
            
            try:
                img_array = cv2.imread(os.path.join(path, img))
                new_img_array = cv2.resize(img_array, (img_size, img_size))
                training_data.append([new_img_array, class_num])
                total_images += 1
                
                # Predict the class label using the current model
                prediction = np.argmax(my_model.predict(np.expand_dims(new_img_array, axis=0),verbose=0))
                if prediction == class_num:
                    correct_predictions += 1
                    
                image_counter += 1
            except Exception as e:
                pass

create_testing_data()

accuracy = correct_predictions / total_images
print("Accuracy of the current model:", accuracy)
```

```python
img = cv2.imread('Happy.jpg')
faces = face_cascade.detectMultiScale(img, 1.3, 5)
(x,y,w,h) = faces[0]
face_img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(face_img)
```

```python
img = cv2.imread('sad.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert frame to grayscale
faces = face_cascade.detectMultiScale(gray, 1.3, 4)
for (x, y, w, h) in faces:
    face_img = cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex, ey), (ex+ew, ey+eh), (0, 255, 0), 2)

    # Perform prediction
    face_roi = cv2.resize(roi_gray, (224, 224))
    face_roi = np.expand_dims(face_roi, axis=0)  # Add batch dimension
    face_roi = np.expand_dims(face_roi, axis=-1)  # Add channel dimension
    face_roi = face_roi / 255.0
    face_roi = np.repeat(face_roi, 3, axis=-1)  # Convert to 3-channel image

    predictions = my_model.predict(face_roi)
    emotion_label = emotion_labels[np.argmax(predictions)]

    # Draw emotion label on the image
    cv2.putText(img, emotion_label, (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 0, 255), 2)

plt.figure()
plt.imshow(img, cmap='gray')
plt.show()
```

```python
font_scale=1.5
font=cv2.FONT_HERSHEY_PLAIN
#set rect background to white
rectangle_bgr=(255,255,255)
# make a black img
img=np.zeros((500,500))
#set some text
text="Some text!"
# get the width and height of the text box
(text_width,text_height)=cv2.getTextSize(text,font,fontScale=font_scale,thickness=1)[0]

#text start pos
text_offset_x=10
text_offset_y=img.shape[0]-25

#make the coords of the box with a small padding of two pix
box_coords=((text_offset_x,text_offset_y),(text_offset_x+text_width+2,text_offset_y-text_height-2,))
cv2.rectangle(img,box_coords[0],box_coords[1],rectangle_bgr,cv2.FILLED)
cv2.putText(img,text,(text_offset_x,text_offset_y),font,fontScale=font_scale,color=(0,0,0),thickness=1)

cap=cv2.VideoCapture(-1) #1
# Check if the webcam is opened
if not cap.isOpened():
    cap = cv2.VideoCapture(0)

# Check if the video capture is successful
if not cap.isOpened():
    raise IOError("Cannot open webcam")

while True:
    ret,frame=cap.read()
    if not ret:
        break
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cascade.detectMultiScale(gray,1.1,4)
    
    for x, y, w, h in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = frame[y:y+h, x:x+w]
        cv2.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        faces=face_cascade.detectMultiScale(roi_gray)
        if len(faces)!=0:
            for (ex,ey,ew,eh) in faces:
                face_roi=roi_color[ey:ey+eh,ex:ex+ew]
                final_image=cv2.resize(face_roi,(224,244))
                final_image=np.expand_dims(final_image,axis=0)
                final_image=final_image/255.0
                
                font=cv2.FONT_HERSHEY_SIMPLEX
                
                predictions=my_model.predict(final_image)
                font_scale=1.5
                font=cv2.FONT_HERSHEY_PLAIN
                
                
                if np.argmax(predictions) in emotion_labels:
                    emotion_label = emotion_labels[np.argmax(predictions)]
                    status = emotion_label
                
                    x1, y1, w1, h1 = 0, 0, 175, 75
                    cv2.rectangle(frame, (x1, x1), (x1 + w1, y1 + h1), (0, 0, 0), -1)
                    cv2.putText(frame, status, (x1 + int(w1 / 10), y1 + int(h1 / 2)), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)
                
    cv2.imshow('Video', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
